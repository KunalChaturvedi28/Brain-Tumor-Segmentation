```python
import numpy as np
import cv2
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from time import time
import glob
```

```python
def fuzzy_segment(image):
    img = cv2.imread(image)
    img = cv2.medianBlur(img, 3)
    m,n,l=img.shape
    Z=img.reshape((-1,3))
    Z = np.float32(Z)
    new_time = time()
    cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans( Z.T, 4, 2, error=0.005, maxiter=1000, init=None)
    print('Fuzzy time for cluster')
    print(time() - new_time,'seconds')
    vpc=(sum(sum(x**2 for x in u)))/(m*n);
    vpe=-((sum(sum(x*np.log2(x) for x in u)))/(m*n));
    print('partition coefficient Vpc = \n', vpc);
    print('partition entropy Vpe= \n',vpe);
    maxu = np.max(u, axis=0)
    crispu=np.zeros(u.shape)
    for i in range(4):
        for j in range(len(Z)):
            if(u[i,j]==maxu[j]):
                crispu[i,j]=1
    img1=crispu[0,:].reshape(m,n)
    img2=crispu[1,:].reshape(m,n)
    img3=crispu[2,:].reshape(m,n)
    img4=crispu[3,:].reshape(m,n)
    fig = plt.figure()
    plt.figure(figsize=(12,12))

    plt.subplot(1, 4, 1)
    plt.imshow(img1)

    plt.subplot(1, 4, 2)
    plt.imshow(img2)

    plt.subplot(1, 4, 3)
    plt.imshow(img3)

    plt.subplot(1, 4, 4)
    plt.imshow(img4)
    plt.show()
    return
```

```python
for file in glob.glob(r"C:\Users\KULDEEP CHATURVEDI\dataset\*.bmp"):
    fuzzy_segment(file)
```
